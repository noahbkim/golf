```python
%matplotlib widget

plt.close()

lift = np.polynomial.Polynomial((0.05, 1, -0.5))
x = np.linspace(0, 1)
y = lift(x)
points = [(0.05, 0.1), (0.08, 0.125), (0.077, 0.14), (0.1, 0.14), (0.18, 0.251), (0.31, 0.3)]

plt.plot(x, y)
plt.scatter([p[0] for p in points], [p[1] for p in points])
plt.grid()
plt.show()
```
